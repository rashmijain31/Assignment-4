```python
df = df.fillna({'Reviews_Per_Month':0})
```

```python
#Bar Chart - Distribution of Room Type:

Room_Type_counts = df['Room_Type'].value_counts()

plt.figure(figsize=(10, 5))
plt.bar(Room_Type_counts.index, Room_Type_counts.values, color='blue', width=0.3)
plt.xlabel('Room Type')
plt.ylabel('Frequency')
plt.title('Room Type Frequencies')
plt.show()
```

```python
#Histogram - Distribution of Price

plt.hist(df['Price'], bins=15, color='lightblue', edgecolor='blue')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Airbnb Listing basis Price')
plt.show()
```

```python
#Scatter Plot - Price vs. Number of Reviews:

plt.scatter( df['Price'], df['Number_Of_Reviews'], alpha=0.5, color='red')
plt.xlabel('Price')
plt.ylabel('Number Of Reviews')
plt.title('Number of Reviews basis Price')
plt.show()
```

```python
#Box Plot

plt.figure(figsize=(10, 5))
sns.boxplot(x='Room_Type', y='Minimum_Nights', data = df)
plt.xlabel('Room Type')
plt.ylabel('Nights Booked')
plt.title('Minimum Stay basis Room Type')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#Pie Chart - Neighborhood Group Distribution:

Neighbourhood_Group = df['Neighbourhood_Group'].value_counts()
plt.figure(figsize=(10, 5))
plt.pie(Neighbourhood_Group.values, labels=Neighbourhood_Group.index, autopct='%1.1f%%')
plt.title('Neighbourhood Group Distribution')
plt.show()
```

```python
# bar chart
job_title_counts = df1['Position'].value_counts()

plt.figure(figsize=(10, 5))
plt.bar(job_title_counts.index, job_title_counts.values)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Position')
plt.ylabel('Count')
plt.title('Job Titles Distribution')
plt.tight_layout()
plt.show()
```

```python
#  pie chart
EmploymentStatus_counts = df1['EmploymentStatus'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(EmploymentStatus_counts, labels=EmploymentStatus_counts.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Active Employees')
plt.show()
```

```python
#Histogram - Distribution of Salary basis Position

plt.hist(df1['Salary'], bins=15, color='orange', edgecolor='black')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Airbnb Listing basis Price')
plt.show()
```

```python
#This visualization will help you understand the relationship between the reasons for termination and 
# the corresponding average salary

#Larger bubbles represent higher average salaries, and the position on the x-axis indicates the reason for termination.

grouped_data = df1.groupby('TermReason')['Salary'].mean().reset_index()

#Set the size of the bubbles based on the average salary
sizes = grouped_data['Salary']

# Set colors for the bubbles
colors = plt.cm.viridis(grouped_data.index / len(grouped_data))

# Plot the bubble chart
plt.scatter(grouped_data.index, grouped_data['Salary'], s=sizes, c=colors, alpha=0.7)

# Add labels and title
plt.xlabel('Reason for Termination')
plt.ylabel('TermReason')
plt.title('HR Data Visualization: Reason for Termination vs. Salary')

# Customize x-axis labels with the TermReason
plt.xticks(grouped_data.index, grouped_data['TermReason'], rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Department', y='Salary', data = df1)
plt.xlabel('Department')
plt.ylabel('Salary')
plt.title('Salary Distribution by Department')
plt.xticks(rotation=45, ha='right')
plt.show()
```
